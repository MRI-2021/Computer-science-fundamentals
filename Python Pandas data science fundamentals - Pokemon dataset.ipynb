```python
## Import datasets - Working with big datasets - chunksize
url = 'https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv'
## My dataframe will be made of just 5 rows. With big data you can chunk 100.000 rows at times.
# for df in pd.read_csv(url, chunksize=5):
#   print(df)

## To compresse big datasets but keeping meaningful information you can groupby and count chunks and then concatenate them in a new dataframe.
new_df = pd.DataFrame(columns=df.columns) ## It creates an empty dataframe with the same columns of the big dataset.

for df in pd.read_csv(url, chunksize=5):
  results=df.groupby(['Type 1']).count()

  new_df = pd.concat([new_df, results])

  print(new_df)
```
